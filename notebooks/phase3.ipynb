```python
import pandas as pd
```

```python
df = pd.read_table('../phase3a/summary.hsq', names=['chr', 'pop1', 'pop2', 'i', 'grm', 'pheno', 'rG', 'se'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfwide = df.query('grm == "kevin"').pivot(index='i', columns='pheno', values='rG')
```

```python
dfwide = dfwide.loc[:,[
    'ALLERGIC.RHINITIS',
    'ASTHMA',
    'CARD',
    'DIA2',
    'DYSLIPID',
    'HYPER',
    'MACDEGEN',
    'OSTIOA',
    'OSTIOP'
]]
```

```python
dfwide.columns = [
    'Allergic Rhinitis',
    'Asthma',
    'Cardiovascular Disease',
    'Type 2 Diabetes',
    'Dyslipidemia',
    'Hypertension',
    'Macular Degeneration',
    'Ostioporosis',
    'Ostioarthritis'
]
```

```python
ax = dfwide.boxplot(return_type='axes', figsize=(20, 10))
ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
plt.show()
```
